Enrichissement des donn√©es brutes avec des indicateurs m√©tier pour am√©liorer le mod√®le de ML.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# [LOGIQUE] : Import des donn√©es nettoy√©es
df = pd.read_csv('../data/generated/cafe_data_2024.csv', parse_dates=['date'])
print(df.head())
```

```python
# [LOGIQUE] : Cr√©ation de la feature 'saison'
# [SYNTAXE] : .dt.month permet d'extraire le mois. apply() applique une fonction √† chaque ligne.
# [SENIOR] : Pr√©f√©rer np.where ou map() pour de meilleures performances sur de gros volumes.
df['saison'] = df['date'].dt.month.map(lambda x: '√©t√©' if 4 <= x <= 9 else 'hiver')

# [LOGIQUE] : Cr√©ation de la feature 'panier_moyen'
# [SYNTAXE] : Division simple de deux colonnes (op√©ration vectoris√©e)
df['panier_moyen'] = df['ventes_total'] / df['nb_clients']

# [LOGIQUE] : Cr√©ation de la feature 'pic_journee'
# [SYNTAXE] : qcut permet de d√©couper en quantiles. On d√©finit un 'pic' si le nb_clients est dans les 25% sup√©rieurs.
df['pic_journee'] = (df['nb_clients'] > df['nb_clients'].quantile(0.75)).astype(int)

print("‚úÖ Features cr√©√©es : saison, panier_moyen, pic_journee")
```

```python
# [LOGIQUE] : Sauvegarde du nouveau dataset enrichi
# [SENIOR] : Toujours garder une copie du dataset brut et versionner les datasets transform√©s
df.to_csv('../data/generated/cafe_data_featured.csv', index=False)
print("üíæ Dataset enrichi sauvegard√© dans data/generated/cafe_data_featured.csv")
```
